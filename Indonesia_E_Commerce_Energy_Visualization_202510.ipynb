```python
# Visualization for Conceptual Framework figure
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

fig, ax = plt.subplots(figsize=(10, 2.4))
ax.axis("off")

y = 0.5
box_h = 0.28
box_w = [0.22, 0.28, 0.24, 0.20]
gap = 0.05
arrow_gap = 0.018

x = [0.06]
for i in range(1, 4):
    x.append(x[i-1] + box_w[i-1] + gap)

labels = [
    "Information\n(Internet → Awareness)",
    "Transaction\n(E-commerce → Affordability)",
    "Inclusion\n(Digital Tools → Access)",
    "Renewable\nAdoption"
]
colors = ["white", "lightblue", "royalblue", "lightgreen"]

# boxes
for xi, wi, lab, col in zip(x, box_w, labels, colors):
    rect = FancyBboxPatch((xi, y - box_h/2), wi, box_h,
                          boxstyle="round,pad=0.02,rounding_size=0.02",
                          linewidth=1.2, facecolor=col, edgecolor="gray")
    ax.add_patch(rect)
    ax.text(xi + wi/2, y, lab, ha="center", va="center", fontsize=10, color="black")

from matplotlib.patches import FancyArrowPatch

# proportional bold arrows aligned with box height, shifted right
arrow_kw = dict(
    arrowstyle='Simple,tail_width=0.8,head_width=16,head_length=18',
    color='gray',
    lw=0,
    alpha=0.9
)

# shift arrows slightly right by adjusting both start and end
shift = 0.01  # adjust as needed (0.008-0.012 usually works best)
for i in range(3):
    x_start = x[i] + box_w[i] + arrow_gap + shift
    x_end   = x[i+1] - arrow_gap + shift
    ax.add_patch(FancyArrowPatch((x_start, y), (x_end, y), **arrow_kw))

# little more space to the right
ax.set_xlim(0, 1.2)   # increase from 1.05
ax.set_ylim(0.15, 0.85)

plt.tight_layout()
plt.show()
```
